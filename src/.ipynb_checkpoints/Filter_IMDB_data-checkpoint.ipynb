```python
# Specify data directory

data_dir = os.path.join(os.path.dirname(os.getcwd()),'data')
data_dir
```

```python
file_path = os.path.join(data_dir, "imdb_scraped_less.txt")
```

```python
df = pd.read_json(file_path, orient = 'records')
```

```python
votes = [s.replace(',', '') if s is not None else None for s in df.votes]
```

```python
votes = [int(s) if s is not None else None for s in votes]
```

```python
df['votes_int'] = votes
```

```python
df_filtered = df[(df.users_rating >= 7.0) & (df.votes_int >= 50000)]
```

```python
df_filtered.to_json(os.path.join(data_dir, 'movie_list.json'))
```
